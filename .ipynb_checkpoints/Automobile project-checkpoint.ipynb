```python
df.drop(['index'], axis=1, inplace = True)
```

# Q1. what is the average price and average mileage for convertible cars

```python
avr_price_and_mileage = df.groupby(["body-style"])[["price","average-mileage"]].mean().loc[['convertible']]
avr_price_and_mileage
```

# Q2. Sort cars by company and price in decending order

```python
sort_cars = df.sort_values(by=['company','price'], ascending=False)
pd.DataFrame(sort_cars)
sort_cars
```

# Q3. Save data for 'Toyota' separately


# Q4. which company is producing most expensive car?

```python
most = df.groupby(['company'])["price"].sum()
```

```python
most_expensive_car = most.nlargest(1)
most_expensive_car = pd.DataFrame(most_expensive_car)
most_expensive_car
```

# Q5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower


```python
max_cars = df.groupby(['company'])[['length','wheel-base','horsepower']].max().sort_values(by =['length','wheel-base','horsepower'],ascending=False)
```

```python
max_cars
```

# Q6. company wise average price of cars and order them in decending order


```python
aver_price_cars = df.groupby(['company'])['price'].mean().sort_values(ascending=False).round(decimals=2)
aver_price_cars = pd.DataFrame(aver_price_cars)
aver_price_cars
```

# Q7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"


```python
df['decision'] = np.where((df['wheel-base'] < 100) & (df['length'] < 175), "don't buy", "buy")
```

```python
decision = df
decision
```

# Q8. Select only cars whose decision is "buy"; find average price based on company of car

```python
average_price_by_company = df[df['decision'] == 'buy'].groupby('company')['price'].mean()
average_price_by_company = pd.DataFrame(average_price_by_company)
average_price_by_company
```

```python
writer = pd.ExcelWriter("Automobile_Output_File.xlsx")

avr_price_and_mileage.to_excel(writer, sheet_name='avr_price_and_mileage')
sort_cars.to_excel(writer, sheet_name='sort_cars_by_company_and_price',index=False)
most_expensive_car.to_excel(writer, sheet_name='most_expensive_car')
max_cars.to_excel(writer, sheet_name='max_cars')
aver_price_cars.to_excel(writer, sheet_name='aver_price_cars')
decision.to_excel(writer, sheet_name='decision',index=False)
average_price_by_company.to_excel(writer, sheet_name='average_price_by_company')

writer.close() # The save() method has been deprecated and removed in Pandas. You should use close() instead.
```
